# Análisis de uso de redes sociales

```python
# Seleccionar columnas relevantes
p7_13_cols = ['P7_13_1', 'P7_13_2', 'P7_13_3', 'P7_13_4', 'P7_13_5', 'P7_13_6']
p7_13_labels = {
    'P7_13_1': 'Facebook',
    'P7_13_2': 'Twitter',
    'P7_13_3': 'Instagram',
    'P7_13_4': 'LinkedIn',
    'P7_13_5': 'Snapchat',
    'P7_13_6': 'Otras redes'
}

# Calcular frecuencias
p7_13_data = usuarios_df[p7_13_cols].apply(lambda x: x.replace({2: 0, 'b': None})).astype(float)
p7_13_totals = p7_13_data.sum()

# Crear gráfica de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=p7_13_totals.index.map(p7_13_labels), y=p7_13_totals.values, palette="Purples_d")
plt.title('Uso de Redes Sociales')
plt.xlabel('Red Social')
plt.ylabel('Número de usuarios')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Reemplazar valores en las columnas de P7_13
p7_13_data = usuarios_df.copy()
p7_13_cols = ['P7_13_1', 'P7_13_2', 'P7_13_3', 'P7_13_4', 'P7_13_5', 'P7_13_6']
p7_13_labels = {
    'P7_13_1': 'Facebook',
    'P7_13_2': 'Twitter',
    'P7_13_3': 'Instagram',
    'P7_13_4': 'LinkedIn',
    'P7_13_5': 'Snapchat',
    'P7_13_6': 'Otras redes'
}

# Reemplazar valores: 2 por 0, 'b' o espacios en blanco por NaN, y convertir a float
p7_13_data[p7_13_cols] = p7_13_data[p7_13_cols].replace({2: 0, 'b': None, ' ': None}).astype(float)
```

```python
# Unir los datos de uso de redes sociales con ESTRATO, ENT y SEXO
p7_13_melted = p7_13_data.melt(
    id_vars=['ESTRATO', 'ENT', 'SEXO'],
    value_vars=p7_13_cols,
    var_name='RedSocial',
    value_name='Uso'
)

# Mapear los códigos de redes sociales a sus nombres
p7_13_melted['RedSocial'] = p7_13_melted['RedSocial'].map(p7_13_labels)

# Filtrar los registros donde 'Uso' es 1 (usuarios que usan esa red social)
p7_13_melted = p7_13_melted[p7_13_melted['Uso'] == 1]
```

```python
# Convertir el código de sexo a etiquetas
sexo_labels = {1: 'Hombre', 2: 'Mujer'}
p7_13_melted['SEXO'] = p7_13_melted['SEXO'].map(sexo_labels)

# Crear gráfica de barras
plt.figure(figsize=(12, 6))
sns.countplot(data=p7_13_melted, x='RedSocial', hue='SEXO', palette="Set2")
plt.title('Uso de Redes Sociales por Sexo')
plt.xlabel('Red Social')
plt.ylabel('Número de usuarios')
plt.xticks(rotation=45)
plt.legend(title='Sexo')
plt.tight_layout()
plt.show()
```

```python
# Asignar etiquetas a los estratos si es necesario (ejemplo)
estrato_labels = {1: 'Bajo', 2: 'Medio Bajo', 3: 'Medio', 4: 'Medio Alto', 5: 'Alto'}
p7_13_melted['ESTRATO'] = p7_13_melted['ESTRATO'].map(estrato_labels)

# Crear gráfica de barras
plt.figure(figsize=(12, 6))
sns.countplot(data=p7_13_melted, x='RedSocial', hue='ESTRATO', palette="Accent")
plt.title('Uso de Redes Sociales por Estrato Socioeconómico')
plt.xlabel('Red Social')
plt.ylabel('Número de usuarios')
plt.xticks(rotation=45)
plt.legend(title='Estrato')
plt.tight_layout()
plt.show()
```

```python
# Asignar productos a categorías de tiendas
producto_categoria = {
    'P7_22_1': {'nombre': 'Libros', 'categoria': 'Especializada'},
    'P7_22_2': {'nombre': 'Música y videos', 'categoria': 'Especializada'},
    'P7_22_3': {'nombre': 'Computadoras, laptops o tablets', 'categoria': 'Departamental'},
    'P7_22_4': {'nombre': 'Celulares o accesorios', 'categoria': 'Departamental'},
    'P7_22_5': {'nombre': 'Software', 'categoria': 'Especializada'},
    'P7_22_6': {'nombre': 'Alimentos y bebidas', 'categoria': 'Autoservicio'},
    'P7_22_7': {'nombre': 'Eventos de entretenimiento', 'categoria': 'Especializada'},
    'P7_22_8': {'nombre': 'Productos de viaje', 'categoria': 'Especializada'},
    'P7_22_9': {'nombre': 'Aparatos electrónicos', 'categoria': 'Departamental'},
    'P7_22_10': {'nombre': 'Productos de belleza y cosméticos', 'categoria': 'Autoservicio'},
    'P7_22_11': {'nombre': 'Ropa y accesorios', 'categoria': 'Departamental'},
    'P7_22_12': {'nombre': 'Artículos para el hogar', 'categoria': 'Departamental'},
    'P7_22_13': {'nombre': 'Vehículos o refacciones', 'categoria': 'Especializada'},
    'P7_22_14': {'nombre': 'Productos financieros', 'categoria': 'Especializada'},
    'P7_22_15': {'nombre': 'Consolas y videojuegos', 'categoria': 'Departamental'},
    'P7_22_16': {'nombre': 'Artículos de salud', 'categoria': 'Especializada'},
    'P7_22_17': {'nombre': 'Equipos fotográficos y ópticos', 'categoria': 'Especializada'},
    'P7_22_18': {'nombre': 'Servicio de transporte por plataforma', 'categoria': 'Especializada'},
    'P7_22_19': {'nombre': 'Otro producto', 'categoria': 'Desconocido'}
}
# Crear DataFrame de productos y categorías
import pandas as pd

productos_df = pd.DataFrame.from_dict(producto_categoria, orient='index')
productos_df.reset_index(inplace=True)
productos_df.rename(columns={'index': 'ProductoID'}, inplace=True)

# Calcular frecuencias de los productos
p7_22_data = usuarios_df[p7_22_cols].apply(lambda x: x.replace({2: 0})).astype(float)
p7_22_totals = p7_22_data.sum().reset_index()
p7_22_totals.columns = ['ProductoID', 'TotalUsuarios']

# Unir con el DataFrame de productos y categorías
productos_totales = p7_22_totals.merge(productos_df, on='ProductoID')

# Agrupar por categoría de tienda y sumar los totales
categoria_totales = productos_totales.groupby('categoria')['TotalUsuarios'].sum().reset_index()

# Ordenar las categorías
categoria_totales = categoria_totales[categoria_totales['categoria'] != 'Desconocido']
categoria_totales['categoria'] = pd.Categorical(categoria_totales['categoria'], categories=['Autoservicio', 'Departamental', 'Especializada'], ordered=True)
categoria_totales = categoria_totales.sort_values('categoria')


```
